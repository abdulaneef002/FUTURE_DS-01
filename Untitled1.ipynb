```python
# Install required library (if not already installed)
!pip install textblob wordcloud

# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
from wordcloud import WordCloud

# Load the dataset
file_path = "/content/student_feedback.csv"  # update path if needed
df = pd.read_csv(file_path)

# Drop Unnamed column if present
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Rename columns for readability
df.columns = [
    'Student ID',
    'Well Versed',
    'Explains Clearly',
    'Presentations',
    'Assignment Difficulty',
    'Doubt Solving',
    'Course Structure',
    'Student Support',
    'Recommendation'
]

# ------------------------------------------
# DASHBOARD STARTS HERE
# ------------------------------------------

# 1. 🔍 Overview
print("🎓 Total Responses:", df.shape[0])
print("🧪 Average Recommendation Score:", round(df['Recommendation'].mean(), 2))
print("📘 Most Frequent Rating in 'Explains Clearly':", df['Explains Clearly'].mode()[0])

# 2. 📊 Ratings Distribution – Bar Plot
plt.figure(figsize=(12,6))
df.drop(columns=['Student ID']).mean().sort_values().plot(kind='barh', color='skyblue')
plt.title("📊 Average Ratings per Category")
plt.xlabel("Average Rating")
plt.grid(axis='x')
plt.show()

# 3. 📈 Course Recommendation – Pie Chart
plt.figure(figsize=(6,6))
df['Recommendation'].value_counts().sort_index().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette('coolwarm', 10))
plt.title("📘 Course Recommendation (1-10)")
plt.ylabel("")
plt.show()

# 4. 🔥 Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.drop(columns=['Student ID']).corr(), annot=True, cmap='coolwarm', fmt='.1f')
plt.title("📌 Feature Correlation Heatmap")
plt.show()

# 5. 🧠 Simulated Sentiment Scores for 'Explains Clearly' (as proxy for feedback)
def get_sentiment(score):
    if score >= 8:
        return 'Positive'
    elif score >= 5:
        return 'Neutral'
    else:
        return 'Negative'

df['Sentiment'] = df['Explains Clearly'].apply(get_sentiment)

# Sentiment Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Sentiment', data=df, palette='pastel')
plt.title("🧠 Sentiment Distribution (Simulated from 'Explains Clearly')")
plt.show()

# 6. ☁️ Word Cloud from Column Names (Just for demo if no text feedback)
text = " ".join(col for col in df.columns)
wordcloud = WordCloud(background_color='white', width=800, height=400).generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("☁️ Word Cloud of Survey Fields")
plt.show()

# 7. 💡 Recommendations Based on Averages
print("\n💡 RECOMMENDATIONS:")
if df['Assignment Difficulty'].mean() > 7:
    print("- Consider reducing the difficulty of assignments.")
if df['Doubt Solving'].mean() < 5:
    print("- Improve how doubts are handled post-session.")
if df['Student Support'].mean() < 6:
    print("- Provide more support for students seeking extra help.")
print("- Continue enhancing course structure and clarity.")

```
